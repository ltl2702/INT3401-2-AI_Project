```python
import pandas as pd
import numpy as np
```

```python
df_april = pd.read_csv('/kaggle/input/merge-month/merge_month/merged_04.csv')
df_october = pd.read_csv('/kaggle/input/merge-month/merge_month/merged_10.csv')
```

```python
df_april.shape
```

```python
df_october.shape
```

```python
def clean_data(df):
    df_cleaned = df.drop(columns=['CIN', 'B04B', 'B05B', 'B06B', 'VSB'])
    
    df_cleaned = df_cleaned[(df_cleaned != 9999).all(axis=1)]

    df_cleaned = df_cleaned[~df_cleaned.isin([np.inf, -np.inf]).any(axis=1)]

    return df_cleaned

df_april = clean_data(df_april)
df_october = clean_data(df_october)
```

```python
df_april.shape
```

```python
df_october.shape
```

```python
def fill_missing_with_median(df):
    numeric_columns = df.select_dtypes(include=['number']).columns
    df[numeric_columns] = df[numeric_columns].apply(lambda col: col.fillna(col.median()), axis=0)
    return df

df_april = fill_missing_with_median(df_april)
df_october = fill_missing_with_median(df_october)
```

```python
df_april.to_csv('filled_data_april.csv', index=False)
df_october.to_csv('filled_data_october.csv', index=False)
```
